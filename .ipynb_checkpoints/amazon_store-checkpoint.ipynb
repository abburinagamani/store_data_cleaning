```python
import json
```

```python
#load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data    
```

```python
data=load_data("store_data.json")
```

```python
#cleaning the data 
#data is inconsistent-mixed type(rating)
#missing data(age)
#duplicate data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    for user in data:
        #cleaning data
        raw_rating=str(user["rating"]).strip().lower()
        if (raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating
        #missing values
        raw_age=user.get("age")
        if(raw_age == None):
            user["age"]=None
        unique_users=set()
        cleaned_data=[]
        if(user["name"] in unique_users):  
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
data=clean_data(data)
print(data)
```

```python
#meaningful insights  of data
total_ratings=0
for user in data:
    total_ratings+=float(user["rating"])
print(f"average product rating={total_ratings/len(data)}")
```

```python
poor_ratings = 0
for user in data:
        if(float(user['rating']) < 3.0):
            poor_ratings += 1
print(f"% users with poor rating = {poor_ratings/len(data)*100}%")     
        
        
```
